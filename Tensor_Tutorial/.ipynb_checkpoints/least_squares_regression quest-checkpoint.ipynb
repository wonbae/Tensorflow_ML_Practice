```python
n_feature = 2
PHI = np.ones(X.shape[0]).reshape(-1, 1)
for i in range(1, n_feature):
    PHI = np.concatenate((PHI, np.power(X, i)), axis=1)
print(PHI.shape)
print(PHI[:3])

# shuffle
idx = np.random.permutation(N)
X_ = X[idx]
PHI_ = PHI[idx]
y_ = y[idx]
```
